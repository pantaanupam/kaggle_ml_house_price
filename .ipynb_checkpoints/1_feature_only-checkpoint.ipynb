```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing datasets
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
train_df.drop(['Id'], axis=1, inplace=True)
train_df
```

### finding missing values..

```python
total = train_df.isnull().sum().sort_values(ascending=False)
percent = (train_df.isnull().sum() / train_df.isnull().count()).sort_values(ascending=False)
percent = round(percent * 100, 2)
df_missing = pd.concat([total, percent], axis=1, keys=['total_missing', 'percent_of_missing'])
df_missing.head(15)
```

```python
# dropping features..
train_df = train_df.drop(labels=['PoolQC', 'MiscFeature', 'Alley', 'Fence'], axis=1)
train_df.shape
```

```python
# viewing the correlation in heatmap
corr = train_df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr, linewidths=0.1, cmap='viridis')
plt.show()
```

```python
# viewing only higher correlations
plt.figure(figsize=(15, 10))
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], vmax=1, linewidths=0.1, annot=True, cmap='viridis')
plt.show()
```

```python
features = ['OverallQual', 'YearBuilt', 'YearRemodAdd', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'FullBath', 'TotRmsAbvGrd', 'GarageCars', 'GarageArea', 'SalePrice']
train_df = train_df[features]
train_df
```

#### Visualizing the data

```python
# checking multi colinearity in the featues
corr = train_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr[corr >= 0.6], annot=True, linewidths=0.1, cmap='viridis')
```

```python
features = [x for x in features if x not in ['1stFlrSF', 'TotRmsAbvGrd', 'GarageArea']]
train_df = train_df[features]
```

```python
# visualizing in regression plot
plt.figure(figsize=(20, 20))
for i, feature in enumerate(features):
    plt.subplot(4, 3, i+1)
    sns.regplot(x=train_df[feature], y=train_df["SalePrice"])
```

#### Creating Model for the dataset..

```python
train_df
```

```python
X_train = train_df.iloc[:, :7].values
y_train = train_df.iloc[:, 7].values
```

```python
# using linear regression for model analysis

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
print("Score : ", lin_reg.score(X_train, y_train))
```

```python
# using polynomial regression 

from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=100)
rf_reg.fit(X_train, y_train)
print("Score : ", rf_reg.score(X_train, y_train))
```

```python
# using grid search to find the optimum hyperparameter for randomforest regression..

from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators': [100, 150, 180, 200, 250, 500]}
grid_search = GridSearchCV(estimator=rf_reg, param_grid=parameters, cv=10, n_jobs=-1)
grid_search = grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
# hence using parameter as above..
rf_reg = RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=None,
                      max_features='auto', max_leaf_nodes=None,
                      min_impurity_decrease=0.0, min_impurity_split=None,
                      min_samples_leaf=1, min_samples_split=2,
                      min_weight_fraction_leaf=0.0, n_estimators=200,
                      n_jobs=None, oob_score=False, random_state=None,
                      verbose=0, warm_start=False)
rf_reg.fit(X_train, y_train)
print("Score : ", rf_reg.score(X_train, y_train))
```

### Predicting the test dataset..

```python
test_id = test_df['Id']
test_df.head()
```

```python
features
```

```python
features.remove('SalePrice')
```

```python
test_df = test_df[features]
test_df.isnull().sum()
```

```python
# filling the missing data by mean value..
test_df.fillna(value=test_df.mean(), inplace=True)
```

```python
# using our regressor to predict the test data

X_test = test_df.iloc[:, :7].values
y_pred = rf_reg.predict(X_test)
y_pred
```

```python
op_df = pd.DataFrame({'Id': test_id, 
              'SalePrice': y_pred})
op_df
```
